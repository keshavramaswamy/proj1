```python
feature_dict = {'shop_avg_transactions_count': "select shop_id, count(*)/33.0 as shop_avg_transactions_count  from sales_history where date_purchase < '2015-10-01' GROUP BY shop_id",
               'item_avg_transactions_count': "select item_id, count(*)/33.0 as item_avg_transactions_count from sales_history where date_purchase < '2015-10-01' GROUP BY item_id", 
                'category_avg_transactions_count' : "select item_category_id, count(*)/33.0 as category_avg_transactions_count from sales_history where date_purchase < '2015-10-01' GROUP BY item_category_id",
                'shop_total_sales' : "select shop_id, sum(item_price) as shop_total_sales from sales_history where date_purchase < '2015-10-01' GROUP BY shop_id",
                'item_total_sales' : "select item_id, sum(item_price) as item_total_sales from sales_history where date_purchase < '2015-10-01' GROUP BY item_id",
                'category_total_sales' : "select item_category_id, sum(item_price) as category_total_sales from sales_history where date_purchase < '2015-10-01' GROUP BY item_category_id",
                'shop_days_since_last_purchase':"select shop_id, julianday('2015-09-30') - julianday(max(date_purchase)) as shop_days_since_last_purchase , max(date_purchase) from sales_history where date_purchase < '2015-10-01' group by shop_id",               
                'item_days_since_last_purchase':"select item_id, julianday('2015-09-30') - julianday(max(date_purchase)), max(date_purchase) as item_days_since_last_purchase from sales_history where date_purchase < '2015-10-01' group by item_id",               
                'category_days_since_last_purchase':"select item_category_id, julianday('2015-09-30') - julianday(max(date_purchase)), max(date_purchase) as category_days_since_last_purchase from sales_history where date_purchase < '2015-10-01' group by item_category_id"
               }         
```

```python
result = conn.execute(feature_dict['shop_avg_transactions_count'])
df_shop_avg_transactions_count = pd.DataFrame(result.fetchall())

result = conn.execute(feature_dict['item_avg_transactions_count'])
df_item_avg_transactions_count = pd.DataFrame(result.fetchall())

result = conn.execute(feature_dict['category_avg_transactions_count'])
df_category_avg_transactions_count = pd.DataFrame(result.fetchall())

result = conn.execute(feature_dict['shop_total_sales'])
df_shop_total_sales = pd.DataFrame(result.fetchall())

result = conn.execute(feature_dict['item_total_sales'])
df_item_total_sales = pd.DataFrame(result.fetchall())

result = conn.execute(feature_dict['category_total_sales'])
df_category_total_sales = pd.DataFrame(result.fetchall())

result = conn.execute(feature_dict['shop_days_since_last_purchase'])
df_shop_days_since_last_purchase = pd.DataFrame(result.fetchall())

result = conn.execute(feature_dict['item_days_since_last_purchase'])
df_item_days_since_last_purchase = pd.DataFrame(result.fetchall())

result = conn.execute(feature_dict['category_days_since_last_purchase'])
df_category_days_since_last_purchase = pd.DataFrame(result.fetchall())
```
